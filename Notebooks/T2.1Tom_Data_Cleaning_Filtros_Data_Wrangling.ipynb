#### Método random

```python
import numpy as np
```

```python
#Generar números aleatorios enteros entre 1 y 100

np.random.randint(1,100)
```

```python
#La manera más clásica de generar un número aleatorio, es entre 0 y 1, con decimales.
#No tiene el INT dentro del random

np.random.random()
```

```python
#Genera una lista de N números aleatorios enteros, dentro del intervalo [a, b]
# n = range
    #Parámetros de intervalos
    # a = inicio
    # b = fin


def randint_list (n, a, b):
    x= []
    for i in range(n):
        x.append(np.random.randint(a,b))
    return x
```

```python
#Generar una lista de 25 números, con valores que se encuentren entre 1 y 50
randint_list(25,1,50)
```

# Semilla (Seed)

* Generación de números aleatorios a partir de una semilla
   - Es el número del cual parte la generación de números aleatorios

#### Método semilla de Numpy

```python
#SEED
np.random.seed(2018)
for i in range(6):
    print(np.random.randint(1,100))
    
```
